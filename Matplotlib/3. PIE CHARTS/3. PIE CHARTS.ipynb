## Importing the library

```python
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
from collections import Counter
```

## Pie Charts:

```python
slices = [60,40]
```

#### Basic Pie chart:

```python
plt.pie(slices)
plt.show()
```

#### With labels :

```python
labels = ['Sixty', 'Forty']
plt.pie(slices, labels = labels)
plt.show()
```

#### With Seperators:
- wedge properties

```python
labels = ['Sixty', 'Forty']
plt.pie(slices, labels = labels, wedgeprops= {'edgecolor' : 'black'})
plt.show()
```

#### Change colours of each slices:

```python
slices = [60,40, 100, 20, 75]
labels = ['A', 'B' , 'C' , 'D' , 'E']
colors = ['red' , 'blue', 'green' , 'black' , 'yellow']
plt.pie(slices, labels = labels, colors = colors, wedgeprops= {'edgecolor' : 'black'})
plt.show()
```

HEX colors:

```python
slices = [160,140, 100, 210, 75]
labels = ['A', 'B' , 'C' , 'D' , 'E']
colors = ['#008fd5' , '#fc4f30', '#e5ae37' , '#6d904f' , 'black']
plt.pie(slices, labels = labels, colors = colors, wedgeprops= {'edgecolor' : 'black'})
plt.show()
```

#### Example data:

```python
# Language Popularity
slices = [59219, 55466, 47544, 36443, 35917, 31991, 27097, 23030, 20524, 18523, 18017, 7920, 7331, 7201, 5833]
labels = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java', 'Bash/Shell/PowerShell', 'C#', 'PHP', 'C++', 'TypeScript', 'C', 'Other(s):', 'Ruby', 'Go', 'Assembly']
```

```python
plt.style.use("fivethirtyeight")
plt.pie(slices, labels = labels, wedgeprops = {'edgecolor': 'black'})
plt.title('Programming Language Popularity')
plt.tight_layout()
plt.show()
```

- Pie chart is not usefull when there are so many ROW values to compare

 > For top five programming languages:

```python
# Language Popularity
slices = [59219, 55466, 47544, 36443, 35917]
labels = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java']
```

```python
plt.style.use("fivethirtyeight")
plt.pie(slices, labels = labels, wedgeprops = {'edgecolor': 'black'})
plt.title('Programming Language Popularity')
plt.tight_layout()
plt.show()
```

##### Emphasize a particular slice:

- Explode argument

```python
plt.style.use("fivethirtyeight")
explode = [0,0,0,0.1,0]
plt.pie(slices, labels = labels,explode = explode, wedgeprops = {'edgecolor': 'black'})
plt.title('Programming Language Popularity')
plt.tight_layout()
plt.show()
```

 - Shadow argument:
 Creates 3D pie

```python
plt.style.use("fivethirtyeight")
explode = [0,0,0,0.1,0]
plt.pie(slices, labels = labels,explode = explode, shadow=True, wedgeprops = {'edgecolor': 'black'})
plt.title('Programming Language Popularity')
plt.tight_layout()
plt.show()
```

- Rotate your pie chart :
Play with the starting angle

```python
plt.style.use("fivethirtyeight")
explode = [0,0,0,0.1,0]
plt.pie(slices, labels = labels,explode = explode, shadow=True,startangle = 180, wedgeprops = {'edgecolor': 'black'})
plt.title('Programming Language Popularity')
plt.tight_layout()
plt.show()
```

#### Display the percentage of the slices:

- autopct = "%1.1f%%"

```python
plt.style.use("fivethirtyeight")
explode = [0,0,0,0.1,0]
plt.pie(slices, labels = labels,explode = explode, shadow=True,autopct = "%1.1f%%", wedgeprops = {'edgecolor': 'black'})
plt.title('Programming Language Popularity')
plt.tight_layout()
plt.show()
```
